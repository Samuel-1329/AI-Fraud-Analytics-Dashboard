```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score
)

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("../dataset/creditcard.csv")

df.head()
```

```python
fraud = df['Class'].value_counts()[1]
genuine = df['Class'].value_counts()[0]

print("Fraud Transactions:", fraud)
print("Genuine Transactions:", genuine)

fraud_percent = (fraud / (fraud + genuine)) * 100

print("Fraud Percentage:", fraud_percent)
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")

plt.xlabel("Amount")

plt.show()
```

```python
fraud_amount = df[df['Class'] == 1]['Amount']
genuine_amount = df[df['Class'] == 0]['Amount']

plt.figure(figsize=(10,5))

sns.kdeplot(genuine_amount, label='Genuine', fill=True)
sns.kdeplot(fraud_amount, label='Fraud', fill=True)

plt.title("Fraud vs Genuine Transaction Amounts")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(20,15))

sns.heatmap(df.corr(), cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
X = df.drop('Class', axis=1)

y = df['Class']
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
print(y_resampled.value_counts())
```

```python
sns.countplot(x=y_resampled)

plt.title("Balanced Dataset After SMOTE")

plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled,
    y_resampled,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=50,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    roc_auc_score
)

print("Accuracy Score:")
print(accuracy_score(y_test, rf_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, rf_pred))

print("\nClassification Report:")
print(classification_report(y_test, rf_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(6,4))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Greens'
)

plt.title("Random Forest Confusion Matrix")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```
